## æ­¥é©Ÿ 3: Excel è™•ç†åŠŸèƒ½

```python
def sanitize_folder_name(name: str) -> str:
    """æ¸…ç†è³‡æ–™å¤¾åç¨±ï¼Œç¢ºä¿ç›¸å®¹æ€§"""
    # ç§»é™¤æˆ–æ›¿æ›ä¸å…è¨±çš„å­—å…ƒ
    invalid_chars = ['/', '\\', ':', '*', '?', '"', '<', '>', '|', '#', '%']
    sanitized = name.strip()
    
    for char in invalid_chars:
        sanitized = sanitized.replace(char, '_')
    
    # é™åˆ¶é•·åº¦
    if len(sanitized) > 255:
        sanitized = sanitized[:255].rstrip()
    
    return sanitized

def find_column(worksheet, column_name):
    """åœ¨å·¥ä½œè¡¨ä¸­å°‹æ‰¾æ¬„ä½"""
    for col_idx, cell in enumerate(worksheet[1], start=1):
        if cell.value == column_name:
            return col_idx
    raise ValueError(f"æ‰¾ä¸åˆ° '{column_name}' æ¬„ä½ï¼")

def copy_selected_files(dest_dir, files_to_copy):
    """è¤‡è£½é¸å®šçš„æª”æ¡ˆåˆ°ç›®æ¨™è³‡æ–™å¤¾"""
    copied_files = []
    
    for file_path in files_to_copy:
        if os.path.isfile(file_path):
            try:
                file_name = os.path.basename(file_path)
                dest_path = os.path.join(dest_dir, file_name)
                shutil.copy2(file_path, dest_path)
                copied_files.append(file_name)
            except Exception as e:
                print(f"  âš ï¸ è¤‡è£½æª”æ¡ˆ {file_name} æ™‚ç™¼ç”ŸéŒ¯èª¤: {str(e)}")
    
    return copied_files

print("âœ“ Excel è™•ç†åŠŸèƒ½å·²å°±ç·’")
```

```python
# å…¨åŸŸè®Šæ•¸å’Œè¼”åŠ©å‡½æ•¸
import time
import gc
import pythoncom

# å…¨åŸŸè®Šæ•¸
excel_com_instance = None  # é‡è¦ï¼šå…¨åŸŸ Excel COM å¯¦ä¾‹

def initialize_excel_com():
    """åˆå§‹åŒ–å…¨åŸŸ Excel COM å¯¦ä¾‹"""
    global excel_com_instance
    if WIN32COM_AVAILABLE and excel_com_instance is None:
        try:
            # åˆå§‹åŒ– COM
            pythoncom.CoInitialize()
            
            # å»ºç«‹ Excel å¯¦ä¾‹
            excel_com_instance = win32com.client.Dispatch("Excel.Application")
            excel_com_instance.Visible = False
            excel_com_instance.DisplayAlerts = False
            excel_com_instance.ScreenUpdating = False
            excel_com_instance.EnableEvents = False
            
            print("âœ“ Excel COM å¯¦ä¾‹å·²åˆå§‹åŒ–")
            return True
        except Exception as e:
            print(f"âŒ ç„¡æ³•åˆå§‹åŒ– Excel COM: {str(e)}")
            excel_com_instance = None
            return False
    return excel_com_instance is not None

def cleanup_excel_com():
    """æ¸…ç†å…¨åŸŸ Excel COM å¯¦ä¾‹"""
    global excel_com_instance
    if excel_com_instance is not None:
        try:
            # é—œé–‰æ‰€æœ‰å·¥ä½œç°¿
            while excel_com_instance.Workbooks.Count > 0:
                excel_com_instance.Workbooks(1).Close(False)
            
            # æ¢å¾©è¨­å®š
            excel_com_instance.ScreenUpdating = True
            excel_com_instance.EnableEvents = True
            excel_com_instance.DisplayAlerts = True
            
            # çµæŸ Excel
            excel_com_instance.Quit()
            excel_com_instance = None
            
            # æ¸…ç† COM
            pythoncom.CoUninitialize()
            
            # å¼·åˆ¶åƒåœ¾å›æ”¶
            gc.collect()
            time.sleep(0.5)
            
            print("âœ“ Excel COM å¯¦ä¾‹å·²æ¸…ç†")
        except Exception as e:
            print(f"âš ï¸ æ¸…ç† Excel COM æ™‚ç™¼ç”ŸéŒ¯èª¤: {str(e)}")
            excel_com_instance = None

def restart_excel_com():
    """é‡æ–°å•Ÿå‹• Excel COM å¯¦ä¾‹"""
    cleanup_excel_com()
    time.sleep(1)  # çµ¦ç³»çµ±ä¸€é»æ™‚é–“
    return initialize_excel_com()

def process_reviewer_excel_windows_excel(file_path, reviewer, column_name, output_folder, retry_count=0):
    """
    ä½¿ç”¨ Windows Excel COM è‡ªå‹•åŒ–ï¼ˆä¿®æ­£ç‰ˆ - åªå¥—ç”¨ç¯©é¸ï¼Œä¸åˆªé™¤è³‡æ–™ï¼‰
    
    é‡è¦ç‰¹æ€§ï¼š
    1. ä¸æœƒä¿®æ”¹åŸå§‹æª”æ¡ˆ
    2. ä¸æœƒåˆªé™¤ä»»ä½•è¨˜éŒ„
    3. åªå¥—ç”¨ç¯©é¸è¦–åœ–ä¾†é¡¯ç¤ºç›¸é—œè³‡æ–™
    4. ä¿ç•™æ‰€æœ‰å·¥ä½œè¡¨å’Œè³‡æ–™é©—è­‰
    
    è™•ç†æµç¨‹ï¼š
    1. é–‹å•ŸåŸå§‹æª”æ¡ˆï¼ˆå”¯è®€ï¼‰
    2. ä½¿ç”¨ SaveCopyAs è¤‡è£½å®Œæ•´æª”æ¡ˆåˆ°å­è³‡æ–™å¤¾
    3. åœ¨è¤‡è£½çš„æª”æ¡ˆä¸Šå¥—ç”¨ AutoFilter
    4. AutoFilter åªæœƒéš±è—ä¸ç›¸é—œçš„åˆ—ï¼Œä¸æœƒåˆªé™¤ä»»ä½•è³‡æ–™
    """
    global excel_com_instance
    
    if not WIN32COM_AVAILABLE:
        return False, None, None
    
    # ç¢ºä¿ Excel COM å¯¦ä¾‹å·²åˆå§‹åŒ–
    if excel_com_instance is None:
        if not initialize_excel_com():
            return False, None, None
    
    wb_source = None
    wb_dest = None
    
    try:
        reviewer_name = sanitize_folder_name(str(reviewer).strip())
        reviewer_folder = os.path.join(output_folder, reviewer_name)
        os.makedirs(reviewer_folder, exist_ok=True)
        
        base_name = os.path.basename(file_path)
        name_without_ext = os.path.splitext(base_name)[0]
        new_filename = f"{name_without_ext} - {reviewer_name}.xlsx"
        dst_path = os.path.join(reviewer_folder, new_filename)
        
        # é–‹å•ŸåŸå§‹æª”æ¡ˆï¼ˆå”¯è®€ï¼Œä¸æœƒä¿®æ”¹ï¼‰
        wb_source = excel_com_instance.Workbooks.Open(os.path.abspath(file_path), ReadOnly=True)
        
        # é‡è¦ï¼šä½¿ç”¨ SaveCopyAs ä¿ç•™å®Œæ•´çš„å·¥ä½œç°¿çµæ§‹ï¼ˆåŒ…å«æ‰€æœ‰è³‡æ–™ï¼‰
        wb_source.SaveCopyAs(os.path.abspath(dst_path))
        wb_source.Close(False)
        wb_source = None
        
        print(f"  âœ“ å·²è¤‡è£½å®Œæ•´æª”æ¡ˆåˆ°å­è³‡æ–™å¤¾ï¼ˆä¿ç•™æ‰€æœ‰è³‡æ–™ï¼‰")
        
        # çŸ­æš«å»¶é²ç¢ºä¿æª”æ¡ˆé‡‹æ”¾
        time.sleep(0.2)
        
        # é–‹å•Ÿè¤‡è£½çš„æª”æ¡ˆé€²è¡Œè™•ç†
        wb_dest = excel_com_instance.Workbooks.Open(os.path.abspath(dst_path))
        
        # æ‰¾å‡ºä¸»è¦è³‡æ–™å·¥ä½œè¡¨
        main_ws = None
        data_sheet_names = ['Sheet1', 'Data', 'è³‡æ–™', 'Main', 'ä¸»è¦']
        
        for sheet_name in data_sheet_names:
            try:
                main_ws = wb_dest.Worksheets(sheet_name)
                break
            except:
                continue
        
        if main_ws is None:
            main_ws = wb_dest.Worksheets(1)
        
        print(f"  âœ“ è™•ç†å·¥ä½œè¡¨: {main_ws.Name}")
        
        # å°‹æ‰¾æ¬„ä½
        col_idx = None
        used_range = main_ws.UsedRange
        if used_range and used_range.Rows.Count > 0:
            for col in range(1, used_range.Columns.Count + 1):
                try:
                    if main_ws.Cells(1, col).Value == column_name:
                        col_idx = col
                        break
                except:
                    continue
        
        if not col_idx:
            raise ValueError(f"æ‰¾ä¸åˆ°æ¬„ä½ '{column_name}'")
        
        # é‡è¦ï¼šåªå¥—ç”¨ç¯©é¸ï¼ˆAutoFilterï¼‰ï¼Œä¸åˆªé™¤ä»»ä½•è³‡æ–™
        # AutoFilter æœƒåœ¨ç¯©é¸è¦–åœ–ä¸­éš±è—ä¸ç¬¦åˆæ¢ä»¶çš„åˆ—
        main_ws.AutoFilterMode = False
        used_range.AutoFilter(Field=col_idx, Criteria1=str(reviewer))
        
        print(f"  âœ“ å·²å¥—ç”¨ç¯©é¸æ¢ä»¶ï¼ˆæ‰€æœ‰è³‡æ–™ä»ä¿ç•™åœ¨æª”æ¡ˆä¸­ï¼‰")
        
        # å„²å­˜ä¸¦é—œé–‰
        wb_dest.Save()
        wb_dest.Close()
        wb_dest = None
        
        print(f"  âœ“ è™•ç†å®Œæˆ - æª”æ¡ˆåŒ…å«æ‰€æœ‰è³‡æ–™ï¼Œç¯©é¸è¦–åœ–åªé¡¯ç¤º: {reviewer}")
        return True, reviewer_folder, new_filename
        
    except Exception as e:
        error_msg = str(e)
        print(f"âŒ Excel COM è™•ç†å¤±æ•—: {error_msg}")
        
        # å˜—è©¦é—œé–‰å·¥ä½œç°¿
        for wb in [wb_source, wb_dest]:
            if wb:
                try:
                    wb.Close(False)
                except:
                    pass
        
        # å¦‚æœæ˜¯ COM éŒ¯èª¤ï¼Œå˜—è©¦é‡æ–°å•Ÿå‹• Excel
        if "call was rejected" in error_msg.lower() or retry_count < 2:
            print(f"  âš ï¸ å˜—è©¦é‡æ–°å•Ÿå‹• Excel COM (é‡è©¦ {retry_count + 1}/2)")
            if restart_excel_com():
                return process_reviewer_excel_windows_excel(file_path, reviewer, column_name, output_folder, retry_count + 1)
        
        return False, None, None

print("âœ“ Excel è™•ç†åŠŸèƒ½å·²å°±ç·’ï¼ˆExcel COM æ–¹æ³•ï¼‰")
print("ğŸ“Œ é‡è¦èªªæ˜ï¼š")
print("â€¢ åŸå§‹æª”æ¡ˆä¸æœƒè¢«ä¿®æ”¹")
print("â€¢ è¤‡è£½çš„æª”æ¡ˆåŒ…å«æ‰€æœ‰åŸå§‹è³‡æ–™")
print("â€¢ åªå¥—ç”¨ç¯©é¸è¦–åœ–ï¼Œä¸åˆªé™¤ä»»ä½•è¨˜éŒ„")
print("â€¢ å¯©æŸ¥è€…å¯ä»¥çœ‹åˆ°ç¯©é¸å¾Œçš„è³‡æ–™ï¼Œä½†æ‰€æœ‰è³‡æ–™éƒ½é‚„åœ¨æª”æ¡ˆä¸­")
```

```python
# Power Automate æ•´åˆåŠŸèƒ½ï¼ˆä½¿ç”¨ Premium åŠŸèƒ½ï¼‰
print("\nğŸ”„ Power Automate æ•´åˆé¸é …...")

def create_power_automate_input_file(processed_reviewers, output_folder, sharepoint_info):
    """
    å»ºç«‹ Power Automate å¯è®€å–çš„ Excel æª”æ¡ˆ
    åŒ…å«æ‰€æœ‰éœ€è¦è¨­å®šæ¬Šé™çš„è³‡è¨Š
    """
    if not processed_reviewers:
        return None
    
    try:
        # å»ºç«‹æª”æ¡ˆè·¯å¾‘
        pa_file_path = os.path.join(output_folder, f"PowerAutomate_æ¬Šé™è¨­å®š_{datetime.now().strftime('%Y%m%d_%H%M%S')}.xlsx")
        
        # æº–å‚™è³‡æ–™
        data = []
        for info in processed_reviewers:
            reviewer = info['reviewer']
            folder_name = os.path.basename(info['folder_path'])
            
            # å»ºç«‹è³‡æ–™åˆ—
            data.append({
                'å¯©æŸ¥è€…åç¨±': reviewer,
                'è³‡æ–™å¤¾åç¨±': folder_name,
                'è³‡æ–™å¤¾å®Œæ•´è·¯å¾‘': info['folder_path'],
                'Email': f"{reviewer}@company.com",  # é è¨­æ ¼å¼ï¼Œéœ€è¦ä¿®æ”¹
                'æ¬Šé™ç­‰ç´š': 'Contribute',  # é è¨­ç‚ºåƒèˆ‡æ¬Šé™
                'è™•ç†ç‹€æ…‹': 'å¾…è™•ç†',
                'ç«™å°åç¨±': sharepoint_info.get('site', ''),
                'æ–‡ä»¶åº«': sharepoint_info.get('library', 'Documents')
            })
        
        # å»ºç«‹ DataFrame ä¸¦å„²å­˜ç‚º Excel
        df = pd.DataFrame(data)
        df.to_excel(pa_file_path, index=False, sheet_name='æ¬Šé™è¨­å®šæ¸…å–®')
        
        print(f"\nğŸ“„ å·²å»ºç«‹ Power Automate è¼¸å…¥æª”æ¡ˆ: {os.path.basename(pa_file_path)}")
        return pa_file_path
        
    except Exception as e:
        print(f"âŒ å»ºç«‹ Power Automate æª”æ¡ˆæ™‚ç™¼ç”ŸéŒ¯èª¤: {str(e)}")
        return None

def get_sharepoint_url_from_path(folder_path):
    """å¾è³‡æ–™å¤¾è·¯å¾‘åµæ¸¬ SharePoint è³‡è¨Š"""
    # ç°¡å–®çš„æ¨¡å¼åŒ¹é…ï¼Œå¯¦éš›ä½¿ç”¨æ™‚å¯èƒ½éœ€è¦èª¿æ•´
    sharepoint_patterns = [
        r'.*SharePoint.*sites[/\\]([^/\\]+)',
        r'.*OneDrive.*([^/\\]+)',
    ]
    
    for pattern in sharepoint_patterns:
        match = re.search(pattern, folder_path, re.IGNORECASE)
        if match:
            return {
                'found': True,
                'site': match.group(1),
                'library': 'Documents'
            }
    
    return {'found': False, 'site': '', 'library': 'Documents'}

def open_power_automate():
    """
    é–‹å•Ÿ Power Automate ç¶²é 
    """
    try:
        import webbrowser
        webbrowser.open('https://make.powerautomate.com')
        print("\nğŸŒ å·²é–‹å•Ÿ Power Automate")
        return True
    except:
        return False

print("\nğŸ’¡ Power Automate åˆ†äº«é¸é …ï¼š")
print("1. ç”¢ç”Ÿ Excel è¼¸å…¥æª”æ¡ˆï¼ˆåŒ…å«æ‰€æœ‰å¯©æŸ¥è€…è³‡è¨Šï¼‰")
print("2. å»ºç«‹è©³ç´°çš„æµç¨‹è¨­å®šèªªæ˜")
print("3. ä½¿ç”¨ SharePoint é€£æ¥å™¨è‡ªå‹•æˆæ¬Š")
print("4. ä¸éœ€è¦ Azure AD æ‡‰ç”¨ç¨‹å¼è¨»å†Š")
```
